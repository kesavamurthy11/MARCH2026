```python
#string handson
#example 1
#name in lower case
name="Kesava murthy"
course="Data Engineering"
Area="Anantapur"
```

```python
#lower case of name
lower_Name= name.lower()
print(lower_Name)
```

```python
#uppercase of course
Upper_course= course.upper()
print(Upper_course)
```

```python
#Area in lower course
lower_area= Area.lower()
print(lower_area)
```

```python
#Example for concatination
print(name+" "+course + " "+Area)
```

```python
#Example for repetiton
print(name *10)
```

```python
#length
length_name=len(name)
print(length_name)
```

```python
#length of empty string
empty_len=len("   ")
print(empty_len)
```

```python
#length after concatination
combined= name+" "+course+" "+Area
print(combined)
len(combined)
```

```python
#indexing
#getting first char of name
print(name)
print(name[0])
```

```python
#getting last char of name
print(name[-1])
```

```python
#getting middle char of any number
print(name[7])
```

```python
#slicing
print(Area)
print(Area[1:7])

```

```python
#getting everything except first and last
print(Area[1:-1])
```

```python
#reversing name using slicing
reversed_name=name[::-1]
print(reversed_name)
```
